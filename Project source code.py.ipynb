```python
import PyPDF4
import os
import re
import openpyxl

# Create a new Excel workbook
workbook = openpyxl.Workbook()

# Get the active sheet
sheet = workbook.active

# Define the headings for the columns
headings = ['Registration Number', 'Age', 'Sex', 'Report Date', 'Referral Doctor', 'Name', 'Hemoglobin', 'Hematocrit', 'RBC COUNT', 'MCV', 'MCH', 'MCHC', 'RDW-CV', 'Total Leucocyte Count', 'Neutrophils', 'Lymphocytes', 'Monocytes', 'Eosinophils', 'Basophils', 'Absolute Neutrophil Count', 'Absolute Lymphocyte Count', 'Absolute Monocyte Count', 'Absolute Eosinophil Count', 'Absolute Basophil', 'Platelet Count', 'Mean Platelet Volume']

# Write the headings to the first row
sheet.append(headings)

# Define the patterns for extracting data from the PDFs
patterns = [
    (r'Order\s+(\w{15})', 'Registration Number'),
    (r'Age \/ Sex (\d+)[\sa-z/]*', 'Age'),
    (r'Age \/ Sex \d+ years \/ (\w+)', 'Sex'),
    (r'(?<=Reported)(.*)(?=Referral)','Report Date'),
    (r'Referral Doctor (\w+)', 'Referral Doctor'),
    (r'Name ([\w\s]+) Order', 'Name'),
    (r'Hemoglobin\s+([\d.]+)', 'Hemoglobin'),
    (r'Hematocrit\s+([\d.]+)', 'Hematocrit'),
    (r'HD focussed DC Detection\s+([\d.]+)', 'RBC COUNT'),
    (r'MCV\s*([\d.]+)', 'MCV'),
    (r'MCH\s*([\d.]+)', 'MCH'),
    (r'MCHC\s*([\d.]+)', 'MCHC'),
    (r'RDW-CV\s*([\d.]+)', 'RDW-CV'),
    (r'Total Leucocyte Count\s+([\d.]+)', 'Total Leucocyte Count'),
    (r'Neutrophils\s*([\d.]+)', 'Neutrophils'),
    (r'Lymphocytes\s*([\d.]+)', 'Lymphocytes'),
    (r'Monocytes\s*([\d.]+)', 'Monocytes'),
    (r'Eosinophils\s*([\d.]+)', 'Eosinophils'),
    (r'Basophils\s*([\d.]+)', 'Basophils'),
    (r'Absolute Neutrophil Count\s*([\d.]+)', 'Absolute Neutrophil Count'),
    (r'Absolute Lymphocyte Count\s*([\d.]+)', 'Absolute Lymphocyte Count'),
    (r'Absolute Monocyte Count\s*([\d.]+)', 'Absolute Monocyte Count'),
    (r'Absolute Monocyte Count\s+([\d.]+)', 'Absolute Monocyte Count'),
    (r'Absolute Eosinophil Count\s+([\d.]+)', 'Absolute Eosinophil Count'),
    (r'Absolute Basophil Count\s+([\d.]+)', 'Absolute Basophil'),
    (r'Platelet Count\s+([\d.]+)', 'Platelet Count'),
    (r'Mean Platelet Volume\s+([\d.]+)', 'Mean Platelet Volume')
]

def extract_pdf_data(file_path):
    with open(file_path, 'rb') as f:
        pdf_reader = PyPDF4.PdfFileReader(f,strict=False)
        page = pdf_reader.pages[0]
        text = page.extractText()
        print(type(text))
        text=text.replace('\n','')
        print(text)
        data = {}
        for pattern, key in patterns:
            match = re.search(pattern, text)
            if match:
                data[key] = match.group(1) if len(match.groups()) > 0 else ''
            else:
                data[key] = ''
        print(data)
        return data


pdf_directory = '/content/drive/MyDrive/project'

# Loop over the PDFs in the directory and extract the data from each one
pdf_data = []
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        file_path = os.path.join(pdf_directory, filename)
        data = extract_pdf_data(file_path)
        pdf_data.append(data)

# Combine the data from all PDFs into a single dictionary
info = {}
for data in pdf_data:
    info.update(data)

 # Create a list to hold the values for each row
    row_values = []
    
    # Loop through the headings and append the values from the info dictionary
    for heading in headings:
        value = info.get(heading, '')
        row_values.append(value)
    
    # Write the row to the Excel sheet
    sheet.append(row_values)
    
# Save the Excel file
workbook.save('cbp_analysis.xlsx')
```
